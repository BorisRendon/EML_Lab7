### Vamos a separar los json de las columnas

```python
json_cols = ['genres']
def clean(x):
    '''
    json.loads() method can be used to parse a valid JSON string and
    convert it into a Python Dictionary. It is mainly used for 
    deserializing native string, byte, 
    or byte array which consists of JSON data into Python Dictionary.
    '''
    return json.loads(x)

for x in json_cols:
    movies_meta[x] = movies_meta[x].apply(clean)
```

```python
def new_json2(x):

    ls = []

    for y in range(len(x[0])):

        for k, v in x[0][y].items():
            ls.append(str(k)+ "_" +str(v))
   
    for z in range(len(ls)):

        if ls[z] not in movies_meta.columns:
            movies_meta[ls[z]] = 0
            movies_meta[ls[z]].iloc[x.name] = 1
        else:
            movies_meta[ls[z]].iloc[x.name] = 1
    return

for x in json_cols:
    movies_meta[[x]].apply(new_json2, axis=1)
```

### Funciones para el sistema de recomendación

```python
def cost(Y, X, T):
    return(((X @ T.T - Y) ** 2) * ((Y != 0) * 1)).sum()


def gradient(Y, X, T):
    R = (Y != 0) * 1
    hip_error = (X @ T.T - Y) * R

    return (
        hip_error @ T,
        hip_error.T @ X,
    )


def adam(
        Y,
        Xo,
        To,
        fun,
        jac,
        alpha=0.0001,
        beta1=0.9,
        beta2=0.999,
        epsilon=0.0000001,
        max_iter=1000
):
    xm = np.zeros(Xo.shape)
    tm = np.zeros(To.shape)

    xv = np.zeros(Xo.shape)
    tv = np.zeros(To.shape)

    X, T, t = Xo, To, 0
    cost = []
    

    while t < max_iter:
        t += 1

        xg, tg = jac(Y, X, T)
        #cost.append(fun(Y, X, T).item())
    

        xm = beta1 * xm + (1 - beta1) * xg
        tm = beta1 * tm + (1 - beta1) * tg

        xv = beta2 * xv + (1 - beta2) * xg * xg
        tv = beta2 * tv + (1 - beta2) * tg * tg

        xmh = xm / (1 - beta1 ** t)
        tmh = tm / (1 - beta1 ** t)

        xvh = xv / (1 - beta2 ** t)
        tvh = tv / (1 - beta2 ** t)

        X -= alpha * xmh / (np.sqrt(xvh) + epsilon)
        T -= alpha * tmh / (np.sqrt(tvh) + epsilon)

    return X, T, cost
```
